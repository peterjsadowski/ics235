An example of Gaussian Process regression using GPy.

```
# Need to install this package.
!pip install GPy
```

```
import numpy as np
import GPy

# Define toy 1D data.
X = np.random.uniform(-3.,3.,(20,1))
Y = np.sin(X) + np.random.randn(20,1)*0.05
```

```
# Define model.
kernel = GPy.kern.RBF(input_dim=1, variance=5., lengthscale=1.)
#type GPy.kern.<tab> here:
#GPy.kern.BasisFuncKernel?
model = GPy.models.GPRegression(X,Y,kernel,noise_var=1.0)

from IPython.display import display
display(model)
fig = model.plot()
```

```
# Optimize parameters of the model.
model.optimize(messages=False)
display(model)
fig = model.plot()
```
